```python
df.boxplot(column="Balance", by="Exited")
plt.ylabel("Balance")
plt.show()
```

```python
plt.scatter(df["EstimatedSalary"], df["Balance"])
plt.title("Balance vs EstimatedSalary")
plt.xlabel("EstimatedSalary")
plt.ylabel("Balance")
plt.show()
```

```python
bins = [20, 30, 50, 70, 90]
labels = ["20-29", "30-49", "50-69", "70+"]
sum = df["Age"]
df["age_category"] = pd.cut(df["Age"], bins=bins, labels=labels)

group = df.groupby("age_category")["CreditScore"]

group.plot(
    kind="bar",
)
plt.title("Credit Score by Age")
plt.xlabel("age_category")
plt.ylabel("CreditScore")
plt.xticks(rotation=0)
plt.ticklabel_format(style="plain", axis="y")
plt.tight_layout()
plt.show()
```


The age reveals the customer age spread, showing if more customers lie within younger or older age groups. This shapes targeted product development and outreach by their credit score
