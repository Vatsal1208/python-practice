```python
#  Write a program to check if a list is a palindrome.
def is_palindrome(s):

    left = 0
    right =len(s) - 1
    while left < right:

        if s[left]!=s[right]:
            return False
        
        left += 1
        right -= 1
        
    return True
s = "22"
print(f"The number is palindrome: {is_palindrome(s)}")
```

```python
#Sort a list in ascending order without using the built-in sort() method.
def sort_list(numbers):
    done = False   #Done =Flag;  #Enters loop start as false 
    while not done:
        done = True     # Inside the loop it assumes true when it's done
        i = 0
        while i + 1 < len(numbers): 
          # print(i)
           
           if numbers[i] > numbers[i + 1]:
                numbers[i], numbers[i + 1] = numbers[i + 1], numbers[i]  # If a swap is needed we're not done
                done = False   #flag changed pass needed 
           i = i + 1
    return numbers

my_list = [3, 1, 4, 6, 2]
print("Before:", my_list)
sorted_list = sort_list(my_list)
print("After: ", sorted_list)

sort_list([3, 1, 4, 6, 2])
    
 

```

```python
def compare_adjacent_elements(input_list):
    for i in range(len(input_list) - 1):
        print(f"Comparing {input_list[i]} with {input_list[i + 1]}")
        if input_list[i] == input_list[i + 1]:
         
            print(f"  {input_list[i]} and {input_list[i + 1]} are equal.")
        elif input_list[i] > input_list[i + 1]:
          
            print(f"  {input_list[i]} is greater than {input_list[i + 1]}.")
        else:
            print(f"  {input_list[i]} is less than {input_list[i + 1]}.")

my_list = [4, 7, 2, 9, 1]
compare_adjacent_elements(my_list)
```

```python
# Write a program to find the factorial of a number using a loop
#n * factorial n-1

def factorial (n):
    if n == 0:
        return 1
    else: 
        return n * factorial (n-1)
    
factorial(10)
       
```

```python
# Print all prime numbers between 1 and 50

def is_prime(num):
    if num <=1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True
    
print("Print Prime B/w 1-50: ")
for num in range(1,51):
    if is_prime(num):
        print(num)    
```

```python
def sum_of_digits(n):
    total = 0
    while n > 0:
        total += n%10           #module to remove last digit
        n=n//10             #flood division
    return total

print(sum_of_digits(9999994349394))      
```

```python

def reverse_no(n):
    reverse = 0
    while n > 0:
            digit = n % 10
            reverse = reverse * 10 + digit
            n = n // 10
    return reverse      
try:
    num = int(input("Enter a number: "))
    print("Input no: ",num)
    print("Reversed: ", reverse_no(num))
except ValueError:
    print("Invalid number")
     
#print(reverse_no(1234))  

    
```

```python
def count_vowels(a):
    vowels = "AEIOUaeiou"
    count = 0
    for character in a:
        if character in vowels:
            count += 1
    return count


input_string = input("Enter a string: ")
print("Your Entered string: ",input_string)
print("Number of vowels: ", count_vowels(input_string))
```

```python
'''f(0)= 0
f(1)= 1
f(2)= f(1)+ f(0)
f(n)= f(n-1)+f(n-2)'''

def fibonacci(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
    
for i in range(11):
    print(f"F{i} = {fibonacci(i)}")    
```

```python
def gcd(a, b):
    return a if b == 0 else gcd(b, a % b)


a= int(input("Enter First Number:"))
print("Your First Number:",a)
b= int(input("Enter Second Number:"))
print("Your Second Number:",b)
print("Greatest common divisor:",gcd(a,b))
```
