```python
#  Write a program to check if a list is a palindrome.
def is_palindrome(s):

    left = 0
    right = len(s) - 1
    while left < right:

        if s[left] != s[right]:
            return False

        left += 1
        right -= 1

    return True


s = "tenet"
print(f"The number is palindrome: {is_palindrome(s)}")
```

```python
# Sort a list in ascending order without using the built-in sort() method.
def sort_list(numbers):
    done = False  # Done =Flag;  #Enters loop start as false
    while not done:
        done = True  # Inside the loop it assumes true when it's done
        i = 0
        while i + 1 < len(numbers):
            # print(i)

            if numbers[i] > numbers[i + 1]:
                numbers[i], numbers[i + 1] = (
                    numbers[i + 1],
                    numbers[i],
                )  # If a swap is needed we're not done
                done = False  # flag changed pass needed
            i = i + 1
    return numbers


my_list = [3, 1, 4, 6, 2]
print("Before:", my_list)
sorted_list = sort_list(my_list)
print("After: ", sorted_list)

sort_list([3, 1, 4, 6, 2])
```

```python
def compare_adjacent_elements(input_list):
    for i in range(len(input_list) - 1):
        print(f"Comparing {input_list[i]} with {input_list[i + 1]}")
        if input_list[i] == input_list[i + 1]:

            print(f"  {input_list[i]} and {input_list[i + 1]} are equal.")
        elif input_list[i] > input_list[i + 1]:

            print(f"  {input_list[i]} is greater than {input_list[i + 1]}.")
        else:
            print(f"  {input_list[i]} is less than {input_list[i + 1]}.")


my_list = [4, 7, 2, 9, 1]
compare_adjacent_elements(my_list)
```

```python
# Print all prime numbers between 1 and 50


def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True


print("Print Prime B/w 1-50: ")
for num in range(1, 51):
    if is_prime(num):
        print(num)
```

```python
def sum_of_digits(n):
    total = 0
    while n > 0:
        total += n % 10  # module to remove last digit
        n = n // 10  # flood division
    return total


print(sum_of_digits(9999994349394))
```

```python
def reverse_no(n):
    reverse = 0
    while n > 0:
        digit = n % 10
        reverse = reverse * 10 + digit
        n = n // 10
    return reverse


try:
    num = int(input("Enter a number: "))
    print("Input no: ", num)
    print("Reversed: ", reverse_no(num))
except ValueError:
    print("Invalid number")

# print(reverse_no(1234))
```

```python
"""f(0)= 0
f(1)= 1
f(2)= f(1)+ f(0)
f(n)= f(n-1)+f(n-2)"""


def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)


for i in range(10):
    print(f"F{i} = {fibonacci(i)}")
```

```python
# Write a program to check if a character is a vowel or consonant.


def check_characters(a):
    vowels = "AEIOUaeiou"
    for character in a:
        if character.isalpha():
            if character in vowels:
                print(f"'{character}' is a vowel.")
            else:
                print(f"'{character}' is a consonant.")
        else:
            print(f"'{character}' is not a letter.")


input_string = input("Enter a string: ")
print("Your entered string:", input_string)
check_characters(input_string)
```

```python
"""Write a program to calculate grade based on marks:
- 90+ : A
- 80–89 : B
- 70–79 : C
- 60–69 : D
- Below 60 : Fail"""


def calculate_grade(marks):

    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    elif marks >= 70:
        return "C"
    elif marks >= 60:
        return "D"
    else:
        return "Fail"


try:
    student_marks = float(input("Enter student's marks: "))
    print(f"Student's marks:{student_marks}")

    if 0 <= student_marks <= 100:
        grade = calculate_grade(student_marks)
        print(f"Student's grade is: {grade}")

    else:
        print("Invalid marks. Please enter a value between 0 and 100.")

except ValueError:
    print("Invalid input. Please enter a numerical value for marks.")

except Exception as e:
    print(f"An unexpected error occurred: {e}")

finally:
    print("""
Grading Scale:
 - 90+     : A
 - 80-89  : B
 - 70-79  : C
 - 60-69  : D
 - Below 60 : Fail
""")
```

```python
# Write a program to check if a person is eligible to vote (age ≥ 18).


def check_age(n):

    if n < 0:
        return "Age cannot be negative."
    elif n > 124:
        return "No verified human has lived that long. Please enter a realistic age."
    elif n >= 18:
        return "You are eligible to vote."
    else:
        return "You are not eligible to vote."


try:
    age = int(input("Enter your age: "))
    print(f"Your age is: {age}")
    print(check_age(age))

except ValueError:
    print("Invalid input. Please enter a numerical value for age.")

except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
