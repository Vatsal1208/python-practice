Remove first n characters from a string

```python
def remove(word, a):
    print("Original String", word)
    x = word[a:]
    return x


print("Removing characters from a string")
print(remove("Vatsalmehta", 2))
print(remove("Vatsalmehta", 6))
```

Check Palindrome Number

```python
def is_palindrome(n):

    origional_string = str(n)
    reversed = origional_string[::-1]
    if origional_string == reversed:
        return True
    else:
        return False


print(is_palindrome(121))
print(is_palindrome(1134))
print(is_palindrome("tenet"))
```

Merge two lists using the following condition

Given two lists of numbers, write Python code to create a new list containing odd numbers from the first list and even numbers from the second list.

```python
def combined_list(list1, list2):
    list3 = []
    for i in list1:
        if i % 2 != 0:
            list3.append(i)
    for i in list2:
        if i % 2 == 0:
            list3.append(i)
    return list3


list1 = [10, 20, 25, 30, 35]
list2 = [40, 45, 60, 75, 90]
print("Combined List:", combined_list(list1, list2))
```

Get each digit from a number in the reverse order

```python
def reverse(n):
    new = []
    string = str(n)[::-1]

    for number in string:
        new.append(int(number))

    return new


n = 7536
print(reverse(n))
```
