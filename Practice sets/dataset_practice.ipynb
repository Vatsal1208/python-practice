## Grouping 


```python
sales_by_category = df.groupby("product_category")["total_amount"].sum().reset_index()
print(sales_by_category)
```

```python
orders_by_city = df["city"].value_counts()
print(orders_by_city)
```

```python
avg_rating = df.groupby("product_category")["customer_rating"].mean().reset_index()
avg_rating["customer_rating"] = avg_rating["customer_rating"].round(2)
print(avg_rating)
```

```python
returning_customers_count = df["is_returning_customer"].value_counts()
print(returning_customers_count)
```

```python
female_customers = df[df["gender"] == "Female"]
female_customers
```

```python
unique_genders = df["gender"].nunique()
gender_counts = df["gender"].value_counts()

print(unique_genders)
print("\n", gender_counts)
```

```python
df_sorted = df.sort_values("total_amount", ascending=False)
df_sorted
```

```python
selected_data = df[["customer_id", "product_category", "total_amount"]]
sort_data = selected_data.sort_values("total_amount", ascending=False)
sort_data
```

```python
df["discount_percent"] = (
    df["discount_amount"] / (df["unit_price"] * df["quantity"])
) * 100
print(df["discount_percent"].round(3).value_counts())
```

```python
def age_group(age):
    if age < 30:
        return "Young"
    elif age < 50:
        return "Middle Age"
    else:
        return "Senior"


df["age_group"] = df["age"].apply(age_group)
print(df["age_group"].value_counts())
```

```python
pivot_table = df.pivot_table(
    index="product_category",
    columns="gender",
    values="total_amount",
    aggfunc="sum",
    fill_value=0,
)

print(pivot_table)
```

```python
bins = [18, 30, 50, 80]
labels = ["18-29", "30-49", "50+"]
df["age_category"] = pd.cut(df["age"], bins=bins, labels=labels)

grouped = df.groupby("age_category")["total_amount"].sum()

grouped.plot(kind="bar", figsize=(10, 6))
plt.title("Total Amount Spent By Age")
plt.xlabel("age category")
plt.ylabel("total_amount")
plt.xticks(rotation=0)
plt.ticklabel_format(style="plain", axis="y")
plt.tight_layout()
plt.show()
```
